```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X1)
```

## 2) Choosing the Number of Clusters :

### a- determining n_cluster by using the elbow method : 

```python
kmeans_kwargs = {
    "init": "random",
    "n_init": 5,
    "max_iter": 300,
    "random_state": 42,
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 15), sse)
plt.xticks(range(1, 15))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(range(1, 15), sse, curve="convex", direction="decreasing")
kl.elbow
```

### b- Determining n_cluster by using the silhouette coefficient :

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice we start at 2 clusters for silhouette coefficient
for k in range(2, 13):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python

plt.style.use("fivethirtyeight")
plt.plot(range(2, 13), silhouette_coefficients)
plt.xticks(range(2, 13))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

11 clusters(silhouette coefficient method) VS 6 clusters(elbow method) ?! 

## K-Means Evaluation

```python
from sklearn import metrics
from sklearn.decomposition import PCA
```

```python
# Instantiate k-means and dbscan algorithms
kmeans = KMeans(n_clusters=10 )

# Fit the algorithms to the features
kmeans.fit(scaled_features)

# Compute the silhouette scores for each algorithm
kmeans_silhouette = silhouette_score(scaled_features, kmeans.labels_).round(2)
homogeneity_score = metrics.homogeneity_score(Y, kmeans.labels_).round(2)
completeness_score = metrics.completeness_score(Y, kmeans.labels_).round(2)
v_measure_score = metrics.v_measure_score(Y, kmeans.labels_).round(2)
adjusted_rand_score = metrics.adjusted_rand_score(Y, kmeans.labels_)
adr  = "{:.2f}".format(adjusted_rand_score)
ami = metrics.adjusted_mutual_info_score(Y, kmeans.labels_).round(2)


print('kmeans_silhouette = ',kmeans_silhouette)
print('homogeneity_score = ' ,homogeneity_score)
print('completeness_score = ' , completeness_score)
print('v_measure_score = ', v_measure_score)
print('adjusted_rand_score =  ', adr)
print('adjusted_mutual_info_score = ', ami)
```

# Building a K-Means Clustering Pipeline

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
preprocessor = Pipeline(
    [
        ("scaler", MinMaxScaler()),
        ("pca", PCA(n_components=2, random_state=42)),
    ]
)
```

```python
clusterer = Pipeline(
   [
       (
           "kmeans",
           KMeans(
               n_clusters=10,
               init="random",
               n_init=10,
               max_iter=300,
               random_state=42,
           ),
       ),
   ]
)
```

```python
pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)
```

```python
pipe.fit(X1)
```

```python
predicted_labels = pipe["clusterer"]["kmeans"].labels_
```

```python
preprocessed_data = pipe["preprocessor"].transform(X1)
```

```python
silhouette_score(preprocessed_data, predicted_labels)
```

```python
adjusted_rand_score(Y, predicted_labels)
```

```python
pcadf = pd.DataFrame(
    pipe["preprocessor"].transform(X1),
    columns=["component_1", "component_2"],
)
pcadf["predicted_cluster"] = pipe["clusterer"]["kmeans"].labels_
```

```python

plt.style.use("fivethirtyeight")
plt.figure(figsize=(8, 8))

scat = sns.scatterplot(
    "component_1",
    "component_2",
    s=50,
    data=pcadf,
    hue="predicted_cluster",
    palette="Set2",
)

scat.set_title("Clustering degits")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)

plt.show()
```

```python
datak = pd.read_csv('Emotion_final.csv')
targets1 = list(datak["Emotion"])
corpus1 = list(datak["Text"])
```

```python
# create the transform
Count_vectorizer = CountVectorizer( stop_words=stopwords.words("english"))
# tokenize and build vocab
Count_vectorizer.fit(corpus1)
# summarize
#print(vectorizer.vocabulary_)
# encode document
vector = Count_vectorizer.transform(corpus1)
# summarize encoded vector
print("vector.shape = ",vector.shape)
#print(type(vector))
print(vector.toarray())
```

```python
Trunc = TruncatedSVD(n_components=6, n_iter=7, random_state=42).fit(vector)
kmeans = KMeans(init=Trunc.components_, n_clusters=6, n_init=1)
```

```python
# Fit the algorithms to the features
kmeans.fit(vector)

# Compute the silhouette scores for each algorithm
kmeans_silhouette = silhouette_score(vector, kmeans.labels_)
```
